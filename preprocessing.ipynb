```python
import pandas as pd, numpy as np
```

```python
DATA = "wikidata_attributes_list.csv"
```

```python
df = pd.read_csv(DATA, header = None)
```

```python
df.head()
```

```python
df.columns = ['search term', 'qterm', 'sex or gender', 'date of birth','place of birth','occupation','field of work', 'country of residence', 'religion', 'voice_type', 'ethnic group', 'native language','academic degree']
```

```python
df = df.assign(age = 2017 - pd.to_numeric(df['date of birth'].str[1:5], errors='coerce'))
```

```python
df.drop(['qterm','date of birth','field of work','academic degree'], axis = 1, inplace = True)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
```
