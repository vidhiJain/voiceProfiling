```python
import pandas as pd
FILE_ATTRIBUTES = "attributes.csv"
FILE_LABELS = "labels_with_audio_path.csv"
```

Processing Acoustic Attributes

```python
columns = ['item',
 'duration',
 'meanfreq',
 'sd',
 'median',
 'Q25',
 'Q75',
 'IQR',
 'skew',
 'kurt',
 'sp.ent',
 'sfm',
 'mode',
 'centroid',
 'meanfun',
 'minfun',
 'maxfun',
 'meandom',
 'mindom',
 'maxdom',
 'dfrange',
 'modindx']
```

```python
attributes = pd.read_csv(FILE_ATTRIBUTES, names = columns)
attributes.head()
```

Processing Labels

```python
labels = pd.read_csv(FILE_LABELS)
```

```python
labels.head()
```

Select the label (Improve)
