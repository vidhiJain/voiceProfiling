```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
TRAIN = "voice.csv"
TEST = "voice_test.csv"
```

```python
df = pd.read_csv(TRAIN)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
x_train, y_train = df.iloc[:, :-1].values, [1 if gender == "male" else 0 for gender in df.iloc[:,-1].values]
# x_train.columns
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=5, activation='relu', input_dim=20))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# model.compile(loss=tf.keras.losses.binary_crossentropy,
#               optimizer=tf.keras.optimizers.SGD(lr=0.01, momentum=0.9, nesterov=True))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy']
             )
```

```python
model.fit(x_train, y_train, epochs=2000, batch_size=64)
```

```python
test_df = pd.read_csv(TEST)
test_df = test_df.sample(frac=1).reset_index(drop=True)
x_test, y_test = test_df.iloc[:, :-1].values, [1 if gender == "male" else 0 for gender in test_df.iloc[:,-1].values]
```

```python
model.evaluate(x_test, y_test, batch_size=64)
```
