from tensorflow.contrib.learn.python.learn.datasets.mnist import read_data_sets uses numpy arrays

```python
import numpy as np
import pandas as pd
train = pd.read_csv(TRAIN)
test = pd.read_csv(TEST)
```

```python
test.shape
```

```python
train = train.sample(frac = 1)
train.info()
```

```python
x = tf.placeholder(tf.float64, [None, 20])
x = tf.cast(x, tf.float32)
```

```python
def weight_init(shape):
    return tf.Variable(tf.truncated_normal(shape,stddev = 0.1))
def bias_init(shape):
    return tf.Variable(tf.zeros(shape))
```

```python
W1 = weight_init([20, 10])
b1 = bias_init([10])
```

```python
z1 = tf.matmul(x, W1) + b1
```

```python
a1 = tf.nn.tanh(z1)
```

```python
W2 = weight_init([10, 1])
b2 = bias_init([1])
```

```python
z2 = tf.matmul(a1, W2) + b2
```

a2 = tf.nn.sigmoid(z2)

```python
y = tf.placeholder(tf.float64, [None, 1])
y = tf.cast(y, tf.float32)
```

TODO: Check this cost function

```python
cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels = y, logits = z2))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cost)
```

```python
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(z2, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
index = 0
def next_batch(size):
    global index 
    batch = train.iloc[index:index+size]
    x = batch.iloc[:, :-1].values
#     print(x.shape)
    y = np.asarray([1 if gender=="male" else 0 for gender in batch.iloc[:, -1].values]).reshape((size, 1))
#     print(y.shape)
    return x, y
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(500):
        batch_x, batch_y = next_batch(100)
        if i % 100 == 0:
            train_accuracy = accuracy.eval(feed_dict={
                  x: batch_x, y: batch_y})
            print('step %d, training accuracy %g' % (i, train_accuracy))
        train_step.run(feed_dict={x: batch_x, y: batch_y})
        print('test accuracy %g' % accuracy.eval(feed_dict={
            x: test.iloc[:,:-1], y: np.asarray([1 if gender=="male" else 0 for gender in test.iloc[:, -1].values]).reshape((test.shape[0], 1))}))
```
